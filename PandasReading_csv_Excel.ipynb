AGGREGATE STATISTICS (Groupby)


```python
df = pd.read_csv()
```

```python
df.groupby(['Type 1']).mean().sort_values('HP', ascending=False)  # classificando os HPs e filtrando por media e tipos
```

```python

df['count'] = 1
df.groupby(['Type 1']).count()['count'] # classificar e mostrar um contatdor de classificação na lista 
```

WORKING WITH LARGE AMOUNTS OF DATA

```python
new_df = pd.DataFrame(columns = df.columns)

for df in pd.read_csv('/content/Pokemon.csv', chunksize=5):
  results = df.groupby(['Type 1']).count()

  new_df = pd.concat([new_df, results])
new_df
```
